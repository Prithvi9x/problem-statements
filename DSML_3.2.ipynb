K-Means Algorithm for k = 2 using Manhattan distance

```python
import matplotlib.pyplot as plt
```

```python
points = [(0.1, 0.6), (0.15, 0.71), (0.08, 0.9), (0.16, 0.85), (0.2, 0.3), (0.25, 0.5), (0.24, 0.1), (0.3, 0.2)]


x = [ 0.1, 0.15, 0.08, 0.16, 0.2, 0.25, 0.24, 0.3 ]
y = [ 0.6, 0.71, 0.9, 0.85, 0.3, 0.5, 0.1, 0.2 ]

plt.scatter(x, y)
plt.show()

#points = list(zip(x, y))
centroids = [(0.1, 0.6), (0.3, 0.2)]
```

```python
def euclidean_distance(point1, point2):
    return ((point1[0]-point2[0])**2+(point1[1]-point2[1])**2)**0.5

def get_cluster(points, centroids):
    clusters = {i: [] for i in range(len(centroids))}
    for point in points:
        distance = [euclidean_distance(point, c) for c in centroids]
        closest_index = distance.index(min(distance))
        clusters[closest_index].append(point)
    
    return clusters

def new_centroids_calc(clusters):
    new_centroids = []
    for data in clusters.values():
        if data:
            x_coord = [p[0] for p in data]
            y_coord = [p[1] for p in data]
            new_centroid = (sum(x_coord)/len(data), sum(y_coord)/len(data))  
            new_centroids.append(new_centroid)
    return new_centroids  

        
```

```python
iteration = 0
while True:
    iteration += 1
    print(f"Iteration {iteration}")
    clusters = get_cluster(points, centroids)
    for i, x in clusters.items():
        print(f"cluster {i+1}: {x}")
        
    new_centroids = new_centroids_calc(clusters)
    print("New centroids: ", new_centroids)

    if new_centroids == centroids:
        break
    centroids = new_centroids
```

```python
for i,x in clusters.items():
    print(x)
```

```python
colors = ['red', 'blue', 'yellow']

for i, points in clusters.items():
    x_vals = [p[0] for p in points]
    y_vals = [p[1] for p in points]
    plt.scatter(x_vals, y_vals, color=colors[i], label=f'Cluster {i+1}')

centroid_x = [c[0] for c in centroids]
centroid_y = [c[1] for c in centroids]
plt.scatter(centroid_x, centroid_y, color='black', marker='x', s=100, label='Centroids')


plt.title("K-Means Clustering (2 Clusters)")
plt.legend()
plt.grid(True)
plt.show()
```

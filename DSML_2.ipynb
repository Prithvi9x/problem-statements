```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
filename = "winequality-red.csv"
df = pd.read_csv(filename)
print(df.head())
```

```python
data = df.describe()
print(data)
data.to_csv('wine_description.csv')
```

```python
volatile_data = df['volatile acidity'].describe()
citric_data = df['citric acid'].describe()
alcohol_quantity = df['alcohol'].describe()

print(volatile_data,'\n')
print(citric_data,'\n')
print(alcohol_quantity)
```

```python
for i in df:
    print(f'Range of {i}: {df[i].min()} - {df[i].max()}')
    print(f'Variance of {i}: {df[i].var()} \n')
    print(df[i].describe(),'\n')
```

```python
plt.hist(df['fixed acidity'])
plt.show()
```

```python
for i in df:
    print(f'Histogram for {i} is:')
    plt.hist(df[i])
    plt.show()
    print()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
csv_path = "Datasets/Titanic.csv"
titanic_csv = pd.read_csv(csv_path)

print("From CSV:")
print(titanic_csv.head())
print()
```

```python
df = titanic_csv
print(df.columns)
print()
print(df["Age"].head())
```

```python
subset = df[["Age", "Pclass", "Fare"]]
print(subset.head())
```

```python
print(df.iloc[0:10])
```

```python
female_passengers = df[df["Sex"] == "female"]
print(female_passengers.head())
```

```python
print(df[(df["Sex"] == "female") & (df["Pclass"] == 1)])
```

```python
print(df.loc[10:15, ["Name", "Sex", "Age"]])
```

```python
print(df.sort_values(by="Fare", ascending=False).head())
```

```python
print(df.describe())
print(df.info())
```

```python
print(df.isna().sum())
```

```python
df["Age"].fillna(df["Age"].median())
df["Fare"].fillna(df["Fare"].median())
df["Cabin"].fillna(df["Cabin"].mode()[0])
```

```python
sns.countplot(data=df, x="Survived")
plt.show()
sns.countplot(data=df, x="Sex")
plt.show()
sns.countplot(data=df, x="Pclass")
plt.show()
```

```python
sns.histplot(data=df, x="Age", hue="Survived", kde=True, bins=30)
plt.show()
```

```python
sns.boxplot(data=df, x="Fare", hue="Survived")
plt.show()
```

```python
numeric_cols = ["Survived", "Pclass", "Age", "SibSp", "Parch", "Fare"]
corr = df[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```

```python
sns.pairplot(
    df[["Survived", "Pclass", "Age", "Fare"]],
    hue="Survived",
    diag_kind="kde"
)
plt.suptitle("Pairplot of Selected Features (colored by Survival)", y=1.02)
plt.show()
```

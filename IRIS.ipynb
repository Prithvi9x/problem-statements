```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("IRIS.csv")
df.head()
```

```python
print(list(df.columns))
```

```python
df.dtypes
```

```python
num_col = df.select_dtypes(include='number').columns
cat_col = df.select_dtypes(exclude='number').columns
print(num_col, cat_col)
```

```python
for col in num_col:
    plt.figure(figsize=(6,4))
    plt.hist(df[col], bins=20)
    plt.xlabel(col)
    plt.ylabel("frequency")
    plt.show()
```

```python
for col in num_col:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.xlabel(col)
    plt.show()
```

```python
for col in num_col:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df["species"], y=df[col])
    plt.title(f"{col} by Species")
    plt.show()
```

```python
df[num_col].describe()
```

```python
df[cat_col].describe()
```

```python
ind_data = df[df['species']=='Iris-setosa'][num_col]
ind_data.mean()
```

```python
ind_data.std()
```

```python
ind_data.quantile(0.25)
```

```python
ind_data = df[df['species']=='Iris-virginica'][num_col]
print(ind_data.mean())
print(ind_data.median())
print(ind_data.std())
print(ind_data.quantile(0.75))
```

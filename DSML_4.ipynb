Root Node of Decision Tree Using Gini Index without library

```python
import pandas as pd
from collections import defaultdict
```

```python
filename = "Decision.csv"
df = pd.read_csv(filename)
print(df)
```

```python
def gini_index(attribute):
    total = len(attribute)
    gini_initial = 1.0
    if total == 0:
        return 0
    attribute_count = defaultdict(int)
    for a in attribute:
        attribute_count[a] += 1
    
    for count in attribute_count.values():
        probability = count/total
        gini_initial -= probability**2
    
    return gini_initial
```

```python
def weighted_gini(data, attribute, target):
    total_len = len(data)
    gini_sum = 0.0
    
    for value in data[attribute].unique():
        subset = data[data[attribute] == value]
        labels = subset[target].tolist()
        gini = gini_index(labels)
        gini_sum += (len(labels) / total_len) * gini

    return gini_sum
```

```python
attributes = ['Age', 'Income', 'Gender', 'Marital Status']
gini_values = {}

print("Gini Index for each attribute: ")
for a in attributes:
    gini = weighted_gini(df, a, 'Buys')
    gini_values[a] = gini
    print(f"{a} : {gini:.3f}")
```

```python
root = min(gini_values, key = gini_values.get)
print(f"The minimum Gini Index value is for {root}: {gini_values[root]:.3f}")
print("The root node is: ", root)
```

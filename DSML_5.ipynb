```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("student_performance.csv")
```

```python
plt.figure(figsize=(7,5)) 
sns.scatterplot(x="Math_Score", y="Reading_Score", hue="Gender", data=df, s=100) 
plt.title("Scatter Plot: Math vs Reading Score by Gender") 
plt.show()
```

```python
plt.figure(figsize=(7,5)) 
sns.scatterplot(x="Reading_Score", y="Writing_Score", hue="Gender", data=df, s=100) 
plt.title("Scatter Plot: Reading vs Writing Score by Gender") 
plt.show()
```

```python
plt.figure(figsize=(8,5)) 
sns.barplot(x="Parental_Education", y="Math_Score", data=df) 
plt.title("Bar Plot: Average Math Score by Parental Education") 
plt.xticks(rotation=30) 
plt.show()
```

```python
plt.figure(figsize=(7,5)) 
sns.boxplot(x="Gender", y="Writing_Score", data=df) 
plt.title("Box Plot: Writing Score Distribution by Gender") 
plt.show()
```

```python
plt.figure(figsize=(7,5)) 
plt.hist(df["Age"], bins=5, color="skyblue", edgecolor="black") 
plt.title("Histogram: Age Distribution of Students") 
plt.xlabel("Age") 
plt.ylabel("Frequency") 
plt.show()
```

```python
plt.figure(figsize=(7,5)) 
plt.hist(df["Math_Score"], bins=5, color="skyblue", edgecolor="black") 
plt.title("Histogram: Math_Score Distribution of Students") 
plt.xlabel("Math_Score") 
plt.ylabel("Frequency") 
plt.show()

plt.figure(figsize=(7,5)) 
plt.hist(df["Reading_Score"], bins=5, color="lightgreen", edgecolor="black") 
plt.title("Histogram: Reading_Score Distribution of Students") 
plt.xlabel("Reading_Score") 
plt.ylabel("Frequency") 
plt.show()

plt.figure(figsize=(7,5)) 
plt.hist(df["Writing_Score"], bins=5, color="pink", edgecolor="black") 
plt.title("Histogram: Writing_Score Distribution of Students") 
plt.xlabel("Writing_Score") 
plt.ylabel("Frequency") 
plt.show()
```

```python
plt.figure(figsize=(7,5)) 
plt.hist(df["Parental_Education"], bins=5, color="skyblue", edgecolor="black") 
plt.title("Histogram: Parental_Education Distribution of Students") 
plt.xlabel("Parental_Education") 
plt.ylabel("Frequency") 
plt.show()
```

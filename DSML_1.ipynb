```python
sns.boxplot(df['Age'])
plt.show()
print(df)
```

```python
def removal_box_plot(df, column, low, high):
    removed_outliers = df[((df[column] >= low) & (df[column] <= high)) | (df[column].isna())].copy()

    sns.boxplot(removed_outliers[column])
    plt.show()
    return removed_outliers

outliers = removal_box_plot(df, 'Age', 10, 50)
print(outliers)
```

```python
outliers['Age'].fillna(outliers['Age'].median(), inplace=True)
print(outliers)
```

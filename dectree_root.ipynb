```python
import pandas as pd
from math import log2
```

```python
df = pd.read_csv("Lipstick.csv")
df.head()
```

```python
df.drop("Id", axis=1, inplace=True)
target = "Buys"
attrs = [c for c in df.columns if c!= target]
```

```python
def entropy(counts):
    total = sum(counts)
    return -(sum((c/total)*log2(c/total)for c in counts if c>0))
```

```python
class_counts = df[target].value_counts()
parent = entropy(class_counts.values)

n = len(df)
info_gain = {}
for a in attrs:
    weighted = 0.0
    for val, group in df.groupby(a):
        counts = group[target].value_counts()
        e = entropy(counts.values)
        weighted += len(group)/n*e
    info = parent-weighted
    print(a, "Info gain: ", info)
    info_gain[a] = info
```

```python
root = max(info_gain, key=info_gain.get)
print("\nRoot Node =", root)
```
